```python
#import all required library

import cv2
import matplotlib.pylab as plt
from google.colab.patches import cv2_imshow
from google.colab import files
```

```python
#load the image

def read_file(filename):
  image = cv2.imread(filename)
  cv2_imshow(image)
  return image
uploaded = files.upload()
filename = next(iter(uploaded))
image = read_file(filename)
```

```python
#detect edge of image

dst_sketch, dst_color_sketch = cv2.pencilSketch(image, sigma_s=50, sigma_r=0.07, shade_factor=0.08)
dst_water_color = cv2.stylization(image, sigma_s=50, sigma_r=0.0825)
```

```python
#convert image into Sketch

plt.figure(figsize=(20,14))
plt.subplot(221), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.axis('off'), plt.title('Original Image', size=20)
plt.subplot(222), plt.imshow(dst_sketch, cmap='gray'), plt.axis('off'), plt.title('Pencil-Sketch', size=20)
plt.show()
```

```python
plt.figure(figsize=(20,14))
plt.subplot(221), plt.imshow(cv2.cvtColor(dst_color_sketch, cv2.COLOR_BGR2RGB)), plt.axis('off'), plt.title('Color-Pencil-Sketch', size=20)
plt.subplot(222), plt.imshow(cv2.cvtColor(dst_water_color, cv2.COLOR_BGR2RGB)), plt.axis('off'), plt.title('Stylized-Color', size=20)
plt.show()
```
